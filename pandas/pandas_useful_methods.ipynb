```python
import numpy as np
import pandas as pd
```

```python
tips = pd.read_csv('tips.csv')

tips.head()
```

# Apply on a Single Column

```python
# Create function to apply to dataframe

def cc_last_4(number):
    
    last_4 = str(number)[-4:]
    return int(last_4)
```

```python
# Test function

cc_last_4(1239872398471908237)
```

```python
# Apply function to specific feature for desired results

tips['last_4_cc'] = tips['CC Number'].apply(cc_last_4)

tips.head()
```

# Apply on Multiple Columns

```python
# Write a function that takes multiple features as an argument

def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
# Call the function on the dataframe using the lambda method

tips[['total_bill', 'tip']].apply(lambda tips: quality(tips['total_bill'], tips['tip']), axis = 1)
```

```python
# Assign the output to a new feature

tips['quality'] = tips[['total_bill', 'tip']].apply(lambda tips: quality(tips['total_bill'], tips['tip']), axis = 1)

tips.head()
```

```python
# You may also use .vectorize()

tips['quality'] = np.vectorize(quality)(tips['total_bill'], tips['tip'])

tips.head()
```
